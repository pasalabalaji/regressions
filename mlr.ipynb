```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("Real estate.csv")
house_age=df["X2 house age"] #x1
distance_from_station=df["X3 distance to the nearest MRT station"] #x2

price=df["Y house price of unit area"]
#y=a+b1.x1+b2.x2
# sum y=a n+b1 sum x1+b2 sum x2

# sum y.x1=a sum x1+b1 sum x1^2+b2 sum x1.x2

# sum y.x2=a sum x2+b1 sum x1.x2+b2 sum x2^2

sum_y=sum(price)

yXx1=np.array(price)*np.array(house_age)
sum_yXx1=sum(yXx1)

yXx2=np.array(price)*np.array(distance_from_station)
sum_yXx2=sum(yXx2)

n=len(price)

sum_x1=sum(house_age)
sum_x2=sum(distance_from_station)

x1_2=[i*i for i in price]
x1_2_sum=sum(x1_2)

x1Xx2=np.array(house_age)*np.array(distance_from_station)
x1Xx2_sum=sum(x1Xx2)

x2_2=[i*i for i in distance_from_station]
x2_2_sum=sum(x2_2)

A = np.array([[n,sum_x1,sum_x2],[sum_x1,x1_2_sum,x1Xx2_sum], [sum_x2,x1Xx2_sum,x2_2_sum]])
b = np.array([sum_y,sum_yXx1,sum_yXx2])
x = np.linalg.solve(A, b)
a=x[0]
b1=x[1]
b2=x[2]

print("a :",a," b1 : ",b1," b2: ",b2)


in_age=float(input("Enter house age"))
dist=float(input("enter distance "))

print(" price : ",46.24886477029812-(0.02276030362440818*in_age)-0.0072567876499127345*dist)
```
